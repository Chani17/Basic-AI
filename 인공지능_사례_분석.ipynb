###<언어 분야 : 챗봇(ChatBot)>
- 인공지능 기반의 챗봇(ChatBot)은 채팅(Chatting)과 로봇(Robot)의 합성어로 사람과의 문자 대화를 통해 질문에 알맞은 답이나 관련된 정보를 제공하는 인공지능(AI) 기반의 커뮤니케이션 소프트웨어를 지칭한다. 즉, 채팅하는 로봇을 말한다. 
- 챗봇의 핵심적인 기술은 도형, 문자, 음성 등의 패턴들을 인식하여 인간이 쓰는 언어 즉, 자연어를 처리할 수 있어야 한다. 또한 사용자가 요구하는 정보를 비정형 데이터에서 찾아낼 수 있어야 하고, 현실의 상황을 정보화하고 활용하는 지능화된 서비스를 제공할 수 있어야 한다. 
- 챗봇에서 쓰이는 기술은 크게 다섯가지가 있다.

  1. 패턴인식(Pattern Recognition) →
기계에 의하여 도형, 문자, 음성 등을 식별시키는 것이다.
  2. 자연어 처리(Natural Language Processing) →
인간이 쓰는 언어를 컴퓨터에 인식시켜 처리하는 일로 정보검색, 질의응답, 시스템 자동번역, 통역 등이 포함된다.
  3. 시멘틱 웹(Semantic Web) → 
컴퓨터가 정보자원의 뜻을 이해하고, 논리적 추론까지 할 수 있는 차세대 지능형 웹이다.
  4. 텍스트 마이닝(Text Mining) → 
비정형 텍스트 데이터에서 새롭고 유용한 정보를 찾아내는 과정 또는 기술이다.
  5. 상황인식컴퓨터(Text Aware Computing) → 
가상공간에서 현실의 상황을 정보화하고, 이를 활용하여 사용자 중심의 지능화 된 서비스를 제공하는 기술이다.
- ex_1 : 법무부의 '인공지능 기반 대화형 생활법률 지식 서비스'

  쉽고 재미있게 생활 법률 서비스를 제공하기 위해 대중적인 메신저인 카카오톡이나 라인과 같은 매체를 통해 인공지능 기반 대화형 법률지식 서비스를 제공한다. 부동산 거래, 상속, 교통사고, 임금퇴직금 등 생활 법률 지능정보 서비스를 언제 어디서나 제공함으로서 일상에서의 준법의식을 높인다.
- ex_2 : '부산 모아'

  '부산 모아'는 부산의 정보들을 속속 알려주는 채팅로봇이다. 남포동이나 해운대 맛집 지도는 물론 부산의 날씨나 미세먼지 정보도 간략한 질문만 하면 바로 답을 얻을 수 있다. 또한 지하철이나 버스의 도착 시간도 채팅을 통해 바로 알려준다.


###<음성 분야 : 음성인식 AI비서>
- 음성인식 가상비서(Voice Assistant)는 스마트폰의 서비스 중 하나인 시리(Siri)로 시작되었다. 최근에는 음성인식 AI 비서로 스마트 스피커를 출시하고 있다. 국내에서는 KT가 기가지니를, 삼성전자는 빅스비(Bixby)를 출시하였다.
- 음성인식 AI비서를 제공하기 위한 핵심기술로는 3가지가 있다. 

  1. ASR(Automatic Speech Recognition) → 
  ASR은 발화자의 명령어를 컴퓨터가 이해하는 언어(Text)로 자동 변화하는 기술로 인공신경망 엔진을 활용하여 단어와 문장을 학습한다.
  2. NLU(Natural Language Understanding) →
  NLU는 주어진 Text를 어떤 의미인지를 파악하는 기술이다. "기분 좋은 음악 들려줘" 하면 '기분 좋은 노래'로 지정되어있는 음악을 찾아 틀어달라는 의미로 분석하는 것이다. 발화자의 의도를 파악하기 위하여 다양한 유사패턴을 학습한다.
  3. TTS(Text to Speech)
  TTs는 Text 문장을 음성으로 변화하는 기술이다. 사람이 말하듯이 속도나 높낮이 조절하는 것이 기술력이다. 각각의 기술에 서비스를 구현시키기 위한 다양한 데이터가 유기적으로 결합되어 음성인식 AI비서 서비스가 가능해진다.

  →  음성인식 AI비서는 인공지능 알고리즘과 빅데이터에 기반하여 음성인식(ASR), 음성이해(NLU), 음성합성(TTS) 기술의 발달로 인간과 소통하는 수준에 도달했다. 



###<이미지 분야 : 인공지능으로 알아보는 닮은 연예인 찾기 NUGOO>
- 인공지능으로 알아보는 닮은 연예인 찾기 NUGOO는 deep + lab에서 인공지능 연구 및 분석을 목적으로 제작된 앱이다. 
- 이 앱은 우리나라 연예인 약 380여명에 대해 8만 5천 여장의 사진을 DNN, CNN(Convolutional Neural Network) 알고리즘으로 학습시켜 탄생하게 되었다고 한다.
- 인공지능 모델(NUGOO)의 학습을 위한 과정은 다음과 같다.
  
   얼굴인식(Detect) → 얼굴변환(Transform) & 자르기(Crop) → 수치화(Representation) → 이미지 학습

  한 때, 유행했던 닮은 연예인 찾기 앱은 이와 같은 과정을 거쳐 닮은 연예인을 찾아주는 것이다. 그리고 연예인 얼굴과 닮은 지수를 분석해 해당 연예인과 얼마나 닮았는지를 퍼센트로 나타내준다.


###<자율주행 분야 : 인천공항 안내로봇 '에어스타(AIRSTAR)'>
- 인천공항 안내로봇인 '에어스타'는 자율주행 기능과 인공지능 등 각종 첨단 ICT 기술이 접목된 안내로봇으로 세계 최초의 상용화된 공항 안내 로봇이다. 

- 자율주행 기술은 크게 인지, 판단, 제어 부분으로 분류가 되는데, AI 기술은 전 영역에서 실시간 수집된 외부 정보를 바탕으로 분석, 판단하여 자율주행을 수행한다. 
  - 인지 : 사물, 보행자, 길, 장애물 등의 데이터를 수집하여 주행환경을 인지하는 기술
  - 판단 : 주행환경에 따른 주행상황을 인식하고 최적의 주행조건(경로, 속도 등)을 결정하는 기술
  - 제어 : 주행 및 움직임과 관련된 구동계 등을 제어하는 기술

  → 자율주행의 전 분야에 활용될 수 있겠지만, 현재로서는 인지분야에 주로 활용되고 있다.  

- 인공지능 기술로는 기계학습과 딥러닝이 이용된다.

  - 자율주행을 위해 가장 중요한 AI분야는 시각(비전)이다. 자율주행 분야는 시간 변화에 따른 환경이 복잡하고 불확실성이 높기 때문에 안정적인 서비스를 제공하기 위해서는 다양한 경우에 대한 학습이 요구된다. 이를 테스트 하기 위해서는 참값 데이터 세트가 필요하다. 자율주행 학습용 데이터 세트는 주행 영상과 영상에서의 사람이나 사물 위치와 속성에 대한 참값(ground truth) 정보를 포함한다. 
  - 영상으로부터 객체의 위치를 검출하기 위해 주로 딥러닝(DNN:Deep Neural Network)을 사용한다. 또한 주변 사물의 위치 정보와 센서정보를 융합, 주변의 미래 위치 정보를 예측하는 DRNN(Deep Recurrent Neural Network)등이 연구중이다.
- 여러가지 테스트를 통해 레코드와 사용자의 만족도를 높일 수 있는 데이터를 수집하여 다양한 환경에서 발생될 수 있는 돌발상황 대응을 위한 사전학습을 한다. 

- 딥러닝 기반의 인지/판단 모델 생성 및 훈련이 필요하고 딥러닝 기술로는 실제 주행에서 발생한 사고들과 시행착오 등으로 데이터의 반복 입력을 통한 AI는 자가 학습이 지속되도록 한다. 
- 학습을 통해 강화학습(Reinforcement Learning)이 적용되어 로봇이 스스로 규칙 터득할 수 있도록 한다.
- 강화학습은 인간의 개입 없이도 반복학습을 통해 인공지능이 스스로 목적을 달성하는 과정을 터득해 내는 방법이다. 인공지능은 수십, 수백만 번의 시행착오를 반복하며 보상값을 극대화 하면서 목적을 달성할 수 있는 방법을 스스로 찾아낸다.  


▶ 이러한 기술들 중 가장 흥미를 끌었던 기술은 인공지능의 이미지 분야이다. 이러한 이미지 인식, 분석, 처리 등의 기술을 가지고 현재 의료, 국방 등 다양한 곳에서 쓰일 수 있게 연구중이다. 특히 눈길이 가는 것은 의료 부분에서 환자의 CT촬영이나 x-ray와 같이 검사 사진을 가지고 의사처럼 진단을 내려주는 기계(로봇)를 연구하고 있다는 사실이다. 인공신경망인 CNN(Convolutional Neural Network)을 이용하는데 이 CNN은 뇌에서 영감을 받았다는 사실에 너무 흥미로웠다. 그리고 CNN은 모든 층이 일련의 뉴런으로 이루어져 있다는 사실 또한 너무나 흥미로웠고 신기했으며 궁금한 점들도 꽤나 있었다. 

▶ 이러한 점들을 바탕으로 관련 강의를 찾아들어보기도 했다. 아직은 잘 모르는 부분들도 많도 단어들도 생소하기 때문에 어려운 강의는 이해하기 힘들다. 그렇기에 관련 서적들이나 입문자 수준의 강의를 찾아볼 것이다. 이를 통해 흥미를 잃지 않는다면 포기하지 않고 계속 하게 되므로 흥미를 계속 유지해 나가는 것이 첫 번째 목표다.
